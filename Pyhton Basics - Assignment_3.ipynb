# Question 4:

What is the difference between a function and a function call?

# Answer 4:

A function in lamen terms is some piece of code in which things can be parameterized(if needed) to enhance functionality and computed data vlaues fromt that code of block execution can be returned to the user. This piece of code once written can executed as many time as required with just one line of function call.
In python, function is a type of built-in-class, which is used to instantiate objects which contain this block of code inside it, 0 or more number of parameters associated with that block of code and are able to return something when the code inside the object is executed. If a return statement is encountered during a function call, some appropriate value is returned to user, otherwise None is returned. 

A function call is basically calling the function object to execute the block of code stored inside it. Arguments might or might not be passed in the function call based on its function definition. A function call always returns some value no matter what. Basic syntax for function call is:  
**function_name(comma seperated arguments if any)**

# Question 6:

What happens to variables in a local scope when the function call returns?

# Answer 6:

The local variables are dereferenced from their object assignments when the function call returns, thus we wont be able to access these local variable outside the function execution and the memory of theses objects is reclaimed based on garbage collectors judgment.

# Question 9:

How do you make a function variable refer to the global variable?

# Answer 9:


In the case that a function does reassignment on the variable (which has same name as global variable), we need to tell python that the variable that we are refering to is the global variable itself and not a new local varibale. This can be done by giving a statement before any use of that variable inside the function using global keyword:  
**global global_variable_name**  
In case that there is no reassignment of value for the global variable inside a function and the global variable is just used for reading purpose then we can directly use the global variable name without using the global statement. Python will search for the variable in its local namespace and since there is no reassignment of the variable it wont be able to find any local variable created with same name and thus, python would take the value from global namespace.

# Question 11:

What does the sentence import areallyourpetsnamederic do?

# Answer 11:

This statement is used to import a module named areallyourpetsnamederic inside your main program. When this statement is encountered, python will search for this module inside the system paths mentioned inside sys.path list, these paths will include the directory path of your main program, any user added paths and predefined paths which depend upon the installation of python that python uses to search its installed modules. If python is able to find the module in any of the locations given in sys.path, then it executes the module, otherwise it gives ModuleNotFound error.

# Question 14:

What is the purpose of the try clause? What is the purpose of the except clause?

# Answer 14:

try clause is used to write suspicious block of codes which you expect might give some runtime or logical error. The purpose of the try clause is to execute this block of code, if the block of code executes successfully then well and good, but if some runtime error or logical error occurs in that block of code then catch that error and redirect the program flow to except clause to handle the situation and prevent program from crashing.
except clause is used to write code which will get executed in case try clause catches some error while its execution, so that the program doesn't crash and is able to handle itself.
