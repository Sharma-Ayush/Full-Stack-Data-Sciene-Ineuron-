# Question 2:

What are the three different types of Boolean operators?

# Answer 2:

Three Boolean operators in python are:  
- AND operator: returns True only if both the conditions being evaluated with AND are True.
- OR operator: returns False only if both the conditions being evaluated with OR are False.
- NOT operator: returns True if the condition being evaluated is False and vice versa.

# Question 5:

What are the six comparison operators?

# Answer 5:

Comparison operators compare the values associated with the given operands. Python has six comparison operators and these are given below:  
- a == b : return True if a and b are equal, otherwise False
- a > b : return True if a is greater than b, otherwise False  
- a < b : return True if a is less than b, otherwise False  
- a >= b : return True if a is greater than or equal to b, otherwise False  
- a <= b : return True if a is less than or equal to b, otherwise False  
- a != b : returns True if a is not equal to b, otherwise False

# Question 6:

How do you tell the difference between the equal to and assignment operators?Describe a condition and when you would use one.

# Answer 6:

Assignment operator consists of one = sign. It is used for variable initalization or changing the reference of an existing variable. For example:  
- A = 5 --> assigns new variable A reference to an int object with value as 5.
- A = 6 --> changes the object reference of variable A from int object with value 5 to int object with value 6.

Equal to operator consists of two = signs. It is used for comparing the values associated with two objects or operands. If both the operands have the same value associated with them then it returns True, otherwise False. For example:  
- 5 == 6 --> return False as 5 is not equal to 6.
- A == 5 --> returns True if A is associated with an int or float object containing 5 as its value.

# Question 10:

How can you tell the difference between break and continue?

# Answer 10:

if break is encountered within a loop, then the program exits outs of the loop in the current state and else block, if any available will not be executed.  
If continue is encountered within a loop then it forces the loop to go to its loop definition statement for next iteration and any code within the loop block after continue statement will not be executed. # Question 11:

In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

# Answer 11:

Results wise all will give same results.  

In terms of function call:
- range(10) --> the initial state will take default value of 0 from function definition and step size also will take default value of 1 from function definition.
- range(0,10) --> the initial state will be overwitten by 0 but as default value and this new value is same, it affect the results and step size will take default value of 1 from function definition.
- range(0,10,1) --> the initial state will be overwitten by 0 but as default value and this new value is same and similarly step size also gets overwritten by 1 which is same as its default value. SInce the new values are same as default values it won't affect the results.

# Question 12:

Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop.

# Answer 12:

```python
print("Using for loop:")
for num in range(1,11):
    print(num)
print("Using while loop:")
num = 1
while num<=10:
    print(num)
    num+=1
```
